# Model: DBSCAN & One-Class SVM

```python
import pandas as pd
import numpy as np
from functools import reduce
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.svm import OneClassSVM
from sklearn.metrics import accuracy_score, classification_report

# Step 1: Concatenate the cleaned dataframes back together
train_combined_df = pd.concat(train_dfs)
test_combined_df = pd.concat(test_dfs)
train_combined_df

# Step 2: Implement the cyclic action detection algorithm using SVM and DBSCAN
def detect_cyclic_actions(train_df, test_df):
    # Separate features and labels for training
    X_train = train_df.drop(columns=['time', 'Class'])
    y_train = train_df['Class']
    print("Checkpoint 1")
    # Normalize the features
    scaler = StandardScaler()
    X_train_normalized = scaler.fit_transform(X_train)
    print("Checkpoint 2")

    # Step 2: Apply DBSCAN clustering on the training data
    dbscan = DBSCAN(eps=0.5, min_samples=5)
    train_df['cluster_label'] = dbscan.fit_predict(X_train_normalized)
    print("Checkpoint 3")

    # Step 3: Train the One-Class SVM
    # Filter out the outliers (cluster label = -1)
    X_train_filtered = X_train_normalized[train_df['cluster_label'] != -1]
    print("Checkpoint 4")

    # Create and train the One-Class SVM classifier
    classifier = OneClassSVM(kernel='linear', nu=0.1)
    classifier.fit(X_train_filtered)
    print("Checkpoint 5")

    # Step 4: Test the classifier
    # Separate features and labels for testing
    X_test = test_df.drop(columns=['time', 'Class'])
    y_test = test_df['Class']
    print("Checkpoint 6")

    # Normalize the features for testing
    X_test_normalized = scaler.transform(X_test)
    print("Checkpoint 7")

    # Predict the labels using the trained One-Class SVM classifier
    predicted_labels = classifier.predict(X_test_normalized)
    print("Checkpoint 8")

    # Convert predicted labels (-1 for outliers, 1 for inliers) to binary labels (0 for outliers, 1 for inliers)
    binary_labels = (predicted_labels == 1).astype(int)
    print("Checkpoint 9")

    # Evaluate the classifier
    accuracy = accuracy_score(y_test, binary_labels)
    print(f"Accuracy: {accuracy:.2f}")
    print(classification_report(y_test, binary_labels))

# Step 4: Apply the cyclic action detection algorithm on the combined training and testing dataframes
detect_cyclic_actions(train_combined_df, test_combined_df)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler

def apply_cnn(train_df, test_df):
    # separate features and labels for training and testing
    X_train = train_df.drop(columns=['time', 'Class','cluster_label']).values
    y_train = train_df['Class'].values
    X_test = test_df.drop(columns=['time', 'Class']).values
    y_test = test_df['Class'].values
    print(X_train.shape)
    print(X_test.shape)

    # normalize the features
    scaler = StandardScaler()
    X_train_normalized = scaler.fit_transform(X_train)
    X_test_normalized = scaler.transform(X_test)

    # reshape the input data to be suitable for 1D Convolutional Neural Network
    X_train_cnn = X_train_normalized.reshape(X_train_normalized.shape[0], X_train_normalized.shape[1], 1)
    X_test_cnn = X_test_normalized.reshape(X_test_normalized.shape[0], X_test_normalized.shape[1], 1)

    # create and compile the 1D CNN model
    model = tf.keras.Sequential([
        tf.keras.layers.Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(X_train_normalized.shape[1], 1)),
        tf.keras.layers.MaxPooling1D(pool_size=2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # train the model
    model.fit(X_train_cnn, y_train, epochs=10, batch_size=32, verbose=1)

    # predict the labels on the test data
    y_pred_proba = model.predict(X_test_cnn)
    y_pred = np.round(y_pred_proba).astype(int).flatten()

    # evaluate the classifier
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:.2f}")
    print(classification_report(y_test, y_pred))

# assuming you have the train_combined_df_ds_t (cyclic data for training) and test_combined_df_ds_t (mixture of cyclic and non-cyclic data for testing) DataFrames
apply_cnn(train_combined_df_ds_t, test_combined_df_ds_t)
```
